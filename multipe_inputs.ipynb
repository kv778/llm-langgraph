```python
from typing import Dict, List, TypedDict
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    value: List[int]
    name: str
    operation: str
    result: str
```

```python
def process_values(state: AgentState) -> AgentState:
    """This function handles multiple different inputs"""
    if state['operation'] == '+':
        val = 0
        for v in state['value']:
            val += v
    else:
        val = 1
        for v in state['value']:
            val *= v
    
    state['result'] = state['name'] + f' chose {state["operation"]} which gives {val}!'
    return state
```

```python
graph = StateGraph(AgentState)

graph.add_node("sum", process_values)
graph.set_entry_point("sum")
graph.set_finish_point("sum")
app = graph.compile()
```

```python
answers = app.invoke({'value': [1,2,3,4,5],
                     'name':'kv',
                     'operation': '*'})
print(answers)
```
